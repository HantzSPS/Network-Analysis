```python
#Import Libraries
import networkx as net
import networkx.algorithms.bipartite as bipartite
import matplotlib.pyplot as plot
import scipy
```

```python
#Load the southern women graph data
g = net.davis_southern_women_graph()
```

```python
women = g.graph['top']
clubs = g.graph['bottom']
```
